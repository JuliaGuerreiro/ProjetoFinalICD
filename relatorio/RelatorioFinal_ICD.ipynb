##**Introdução**
O suicídio pode ser definido como um ato deliberado, realizado pelo indivíduo de forma consciente, cuja intenção seja a própria morte, mesmo que ambivalente, usando um meio que ele acredita ser letal. É um comportamento com múltiplos determinantes e resultado de uma complexa interação de fatores psicológicos e biológicos, inclusive genéticos, culturais e socioambientais.

É importante ressaltar que, devido ao preconceito e ao estigma que o suicídio carrega, os dados estatísticos sobre o suicídio são subnotificados, o que mascara uma realidade preocupante, pois estima-se que cerca de um milhão de pessoas morre anualmente, por suicídio, o que levou a OMS a considerar o suicídio como uma epidemia silenciosa. Trata-se de uma questão de saúde pública.

Fonte: https://unilab.edu.br/prevencao-do-suicidio/

Nosso objetivo, com este projeto, sabendo que o suícidio pode resultar de fatores culturais e socioambientais, é analisar as possíveis relações entre o ***Indíce de Desenvolvimento Humano*** de um país e suas taxas de suícidio e investigar este fenômeno.

##**Tratamento de Dados**

Para simplificar a base utilizada neste projeto, o período analisado foi reduzido aos anos de 2010 a 2016. A coluna relativa à idade, que apresentava a idade por intervalos (ex.: 55-74 years) foi removida e em seu lugar foi considerada apenas a geração do conjunto de pessoas. Também, com o intuito de focar apenas em dados específicos, as colunas mantidas para estudo foram: país, ano, sexo, número de suícidios, taxa de suícidio a cada 100k pessoas, IDH, renda per capita e geração (relativa ao intervalo de tempo em que uma determinada pessoa nasceu).

##**Conclusão**

Tendo em vista as nossas perguntas iniciais, podemos concluir alguns pontos. Segundo a análise, ao longo dos anos é possível notar uma mínima variação nas taxas de suicídio quando são considerados o IDH e o GPD de um país. No que foi observado, pode-se dizer que há uma leve tendência em países onde o IDH é mais elevado, no entanto, não pode-se afirmar com exatidão devido aos números baixos retornados.

Um pouco a se ressaltar é aquele mencionado em nossa introdução: os casos de suicídio são normalmente influenciados por uma combinação de fatores e não apenas por um deles isoladamente, e isso pôde ser observado em nossa pesquisa. Isoladamente, os indicadores socio-econômicos foram insuficientes para se chegar a alguma conclusão ou mesmo prever padrões futuros.

Em relação à idade das pessoas, pôde-se observar uma maior têndencia entre as pessoas já adultas. Os casos em crianças e adolescentes foram quase irrelevantes.

Já em relação ao gênero, chegamos a uma conclusão interessante. Pôde-se perceber que os homens possuem uma maior tendência a concluir tal ato do que as mulheres e isso pôde ser observado através de uma rápida análise exploratória e em sua confirmação através de uma regressão. Acreditamos que essa foi uma descoberta interessante mas que também não deve ser considerada isoladamente, tendo em vista que há vários outros fatores que podem vir a influenciar mais fortemente os homens.

Em suma, nota-se que a análise de casos de suicídio é uma análise tão complexa quanto o tema e exige a investigação de vários fatores simultâneamente para que possamos chegar em resultados mais conclusivos e que possam gerar algum impacto positivo na sociedade e na saúde-pública.
